# Kronecker delta (oblique 3D)

Constructs $\delta^i{}_j$ — upper index $i$, lower index $j$ — in the
3D spatial index space with **Oblique** realm, then renders it as LaTeX
both inline (Jupyter rich display) and as a standalone `.tex` file in `out/`.

```python
import pathlib

import tender
from tender import IndexNameMap, Level, Realm
```

```python
sp = tender.space_3d
i = tender.alloc_index()
j = tender.alloc_index()

delta = tender.delta(Realm.Oblique, sp, Level.Upper, Level.Lower, i, j)
```

Jupyter calls `_repr_latex_()` automatically on the last expression in a cell:

```python
delta
```

We can also pass an explicit `IndexNameMap` to control (or inspect) name allocation:

```python
index_map = IndexNameMap()
tex_expr = tender.render_latex(delta, index_map)

print("LaTeX      :", tex_expr)
print("Name for i :", index_map.lookup(i))
print("Name for j :", index_map.lookup(j))
```

Write a standalone LaTeX document to `out/`:

```python
doc = (
    r"\documentclass{article}" "\n"
    r"\usepackage[utf8]{inputenc}" "\n"
    r"\usepackage{amsmath,amssymb}" "\n"
    r"\begin{document}" "\n\n"
    r"\section*{Kronecker delta (oblique 3D)}" "\n\n"
    "The Kronecker delta with one upper and one lower index\n"
    "in the oblique 3D index space:\n"
    r"\[" "\n"
    "  " + tex_expr + "\n"
    r"\]" "\n\n"
    r"\end{document}" "\n"
)

out_dir = pathlib.Path("out")
out_dir.mkdir(exist_ok=True)
out_path = out_dir / "kronecker_delta.tex"
out_path.write_text(doc)

print(f"Written : {out_path}")
print(f"Compile : pdflatex -output-directory out {out_path}")
```
