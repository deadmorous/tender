# A cross-with-identity identity: $a \times I = I \times a$

The identity tensor resolves to the dyad $I = \sum_m e_m \otimes e^m$, so
$a \times I$ is a *vector crossed with a dyad*.  We expand in the World
Cartesian System, distribute the cross over the dyad, turn each
$e_i \times e_m$ into a Levi-Civita term, and canonicalize — and both
$a \times I$ and $I \times a$ reduce to the same thing.

```python
import tender
import tender.basis as tb
import tender.derivation as td
```

## 1. A frame, a vector, and the identity

`@` is the dot, `%` the cross, `*` the tensor product.

```python
ctx = tender.Context()
frame = tb.wcs(ctx)            # orthonormal frame i, j, k
a = tender.tensor("a", rank=1, ctx=ctx)
I = tender.identity(ctx=ctx)
a % I
```

## 2. Expand $I = \sum_m e_m \otimes e^m$

Expanding in the basis turns $a \times I$ into a cross with the identity dyad.

```python
expanded = tb.expand_in_basis(a % I, frame, tb.Variance.Covariant)
expanded
```

## 3. Distribute the cross over the dyad

`distribute_contraction` applies $a \times (u \otimes v) = (a \times u) \otimes v$,
so the cross now acts on the first leg of the dyad.

```python
distributed = td.distribute_contraction(expanded)
distributed
```

## 4. Levi-Civita and canonicalize

`simplify_basis_cross` rewrites $e_i \times e_m \to \varepsilon_{imk} e^k$;
`canonicalize` then normalizes (Levi-Civita symmetry, nested-sum order, ...).

```python
lhs = td.canonicalize(tb.simplify_basis_cross(distributed, frame))
lhs
```

## 5. $I \times a$ reduces to the same thing

Running the identical pipeline on $I \times a$ gives the same coordinate
expression, so $a \times I = I \times a$.

```python
def reduce(e):
    e = tb.expand_in_basis(e, frame, tb.Variance.Covariant)
    e = td.distribute_contraction(e)
    e = tb.simplify_basis_cross(e, frame)
    return td.canonicalize(e)

rhs = reduce(I % a)
print("a×I =", lhs.latex())
print("I×a =", rhs.latex())
td.algebraic_eq(lhs, rhs)
```
